```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.collections import LineCollection
import fastf1 as ff1
from fastf1 import plotting
import plotly.io as pio
from sklearn.preprocessing import OrdinalEncoder

#setting the size for the plots
plt.rcParams["figure.figsize"] = (15,5)
plt.rcParams['figure.dpi'] = 140
plt.rcParams["font.family"] = "serif"


#setting the data to load into the cache
ff1.Cache.enable_cache('F1Cache')
```

```python
#loading data in from session
session = ff1.get_session(2022, 'Bahrain', 'R')
laps = session.load_laps(with_telemetry=True)
lap = laps.pick_fastest()
tel = lap.get_telemetry()

ver = laps.pick_driver('VER')
ham = laps.pick_driver('HAM')
gas = laps.pick_driver('GAS')
lec = laps.pick_driver('LEC')
alo = laps.pick_driver('ALO')
sai = laps.pick_driver('SAI')
oco = laps.pick_driver('OCO')
nor = laps.pick_driver('NOR')
vet = laps.pick_driver('VET')
rai = laps.pick_driver('RAI')
rus = laps.pick_driver('RUS')
gio = laps.pick_driver('GIO')
tsu = laps.pick_driver('TSU')
lat = laps.pick_driver('LAT')
maz = laps.pick_driver('MAZ')
msc = laps.pick_driver('MSC')
ric = laps.pick_driver('RIC')
sto = laps.pick_driver('STR')
bot = laps.pick_driver('BOT')
per = laps.pick_driver('PER')
```

```python
verbest=ver.pick_fastest()
hambest=ham.pick_fastest()
lecbest=lec.pick_fastest()
vertel= verbest.get_telemetry()
hamtel= hambest.get_telemetry()
lectel= lecbest.get_telemetry()
```

```python
plotting.setup_mpl()
fig, ax = plt.subplots()
ax.plot(vertel['Distance'], vertel['Speed'], color='yellow',label='VER')
ax.plot(lectel['Distance'], lectel['Speed'], color='red',label='LEC')
ax.set_title("VER vs LEC 2022 Bahrain Fastest Lap Speed")
ax.set_xlabel("Distance (m)")
ax.set_ylabel("Speed (KM/h)")
plt.legend(loc='lower right')
plt.savefig('F1Plots/MaxVCharlesRaceSpeedBahrain2022.png')
plt.show()
```

```python
plotting.setup_mpl()
fig, ax = plt.subplots()
ax.plot(vertel['Distance'], vertel['Speed'], color='yellow',label='VER')
ax.plot(hamtel['Distance'], hamtel['Speed'], color='cyan',label='HAM')
ax.set_title("VER vs HAM 2022 Bahrain Fastest Lap Speed")
ax.set_xlabel("Distance (m)")
ax.set_ylabel("Speed (KM/h)")
plt.legend(loc='lower right')
plt.savefig('F1Plots/MaxVLewisRaceSpeedBahrain2022.png')
plt.show()
```

```python
plotting.setup_mpl()
fig, ax = plt.subplots()
ax.plot(vertel['Distance'], vertel['Brake'], color='yellow',label='VER')
ax.plot(hamtel['Distance'], hamtel['Brake'], color='cyan',label='HAM')
ax.set_title("VER vs HAM 2022 Bahrain Fastest Lap Braking Points")
ax.set_xlabel("Distance (m)")
ax.set_ylabel("Brake")
plt.legend(loc='lower right')
plt.savefig('F1Plots/MaxVLewisRaceBrakeBahrain2022.png')
plt.show()
```

```python
plotting.setup_mpl()
fig, ax = plt.subplots()
ax.plot(vertel['Distance'], vertel['Throttle'], color='yellow',label='VER')
ax.plot(hamtel['Distance'], hamtel['Throttle'], color='cyan',label='HAM')
ax.set_title("VER vs HAM 2022 Bahrain Fastest Lap Throttle")
ax.set_xlabel("Distance (m)")
ax.set_ylabel("Throttle")
plt.legend(loc='lower right')
plt.savefig('F1Plots/MaxVLewisRaceThrottleBahrain2022.png')
plt.show()
```
