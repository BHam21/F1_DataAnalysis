```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import fastf1 as ff1
from fastf1 import plotting
import time
from datetime import datetime, date, time, timedelta
from timple.timedelta import strftimedelta
from fastf1.core import Laps


#setting parameters for visualizations
plt.rcParams["figure.figsize"] = (15,5)
plt.rcParams['figure.dpi'] = 140
```

```python
#sending the data to the cache
ff1.Cache.enable_cache('F1Cache')
```

```python
#Importing data from the (year,race, qualifying,race, or practice)
race = ff1.get_session(2022, 'Bahrain', 'R')
laps = race.load_laps(with_telemetry=True)
results= race.results

ver = laps.pick_driver('VER')
ham = laps.pick_driver('HAM')
gas = laps.pick_driver('GAS')
lec = laps.pick_driver('LEC')
alo = laps.pick_driver('ALO')
sai = laps.pick_driver('SAI')
oco = laps.pick_driver('OCO')
nor = laps.pick_driver('NOR')
vet = laps.pick_driver('VET')
zho = laps.pick_driver('ZHO')
rus = laps.pick_driver('RUS')
gio = laps.pick_driver('GIO')
tsu = laps.pick_driver('TSU')
lat = laps.pick_driver('LAT')
mag = laps.pick_driver('MAG')
msc = laps.pick_driver('MSC')
ric = laps.pick_driver('RIC')
sto = laps.pick_driver('STR')
bot = laps.pick_driver('BOT')
per = laps.pick_driver('PER')
alb = laps.pick_driver('ALB')
```

```python
#Creating a telemetry plot that shows the race lap times between max and lewis
plotting.setup_mpl()
fig, ax = plt.subplots()
ax.plot(ver['LapNumber'], ver['LapTime'], color='yellow',label='VER')
ax.plot(ham['LapNumber'], ham['LapTime'], color='cyan',label='HAM')
plt.suptitle(f"VER vs HAM Lap Times {race.weekend.name} {race.weekend.year}")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
plt.legend(loc='upper right')
plt.savefig(f'F1Plots/MaxVLewisLapTimes{race.weekend.name}{race.weekend.year}.png')
plt.show()
```

```python
#Plots to show lap time differences between teammates for the race
fig, ax = plt.subplots()
plt.suptitle(f"HAM vs RUS Lap Times {race.weekend.name} {race.weekend.year}")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
ax.plot(ham['LapNumber'], ham['LapTime'], color='cyan',label='HAM')
ax.plot(rus['LapNumber'], rus['LapTime'], color='cyan',linestyle='dashed',label='RUS')
plt.legend(loc='upper right')
plt.savefig('F1Plots/MercedesLapTimes2021Brazil.png')
plt.show()

fig, ax = plt.subplots()
plt.suptitle(f"VER vs PER {race.weekend.name} {race.weekend.year}")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
ax.plot(per['LapNumber'], per['LapTime'], color='yellow',label='PER')
ax.plot(ver['LapNumber'], ver['LapTime'], color='yellow',linestyle='dashed',label='VER')
plt.legend(loc='upper right')
plt.savefig('F1Plots/RedBullLapTimes2022Bahrain.png')
plt.show()

fig, ax = plt.subplots()
plt.suptitle(f"LEC vs SAI {race.weekend.name} {race.weekend.year}")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
ax.plot(lec['LapNumber'], lec['LapTime'], color='red',label='LEC')
ax.plot(sai['LapNumber'], sai['LapTime'], color='red',linestyle='dashed',label='SAI')
plt.legend(loc='upper right')
plt.savefig('F1Plots/FerrariLapTimes2022Bahrain.png')
plt.show()

fig, ax = plt.subplots()
plt.suptitle(f"OCO vs ALO {race.weekend.name} {race.weekend.year}")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
ax.plot(oco['LapNumber'], oco['LapTime'], color='blue',label='OCO')
ax.plot(alo['LapNumber'], alo['LapTime'], color='blue',linestyle='dashed',label='ALO')
plt.legend(loc='upper right')
plt.savefig('F1Plots/AlpineLapTimes2022Bahrain.png')
plt.show()

fig, ax = plt.subplots()
plt.suptitle(f"RIC vs NOR {race.weekend.name} {race.weekend.year}")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
ax.plot(nor['LapNumber'], nor['LapTime'], color='orange',label='NOR')
ax.plot(ric['LapNumber'], ric['LapTime'], color='orange',linestyle='dashed',label='RIC')
plt.legend(loc='upper right')
plt.savefig('F1Plots/McLarenLapTimes2022Bahrain.png')
plt.show()

fig, ax = plt.subplots()
plt.suptitle(f"ZHO vs GIO {race.weekend.name} {race.weekend.year}")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
ax.plot(zho['LapNumber'], zho['LapTime'], color='maroon',label='ZHO')
ax.plot(bot['LapNumber'], bot['LapTime'], color='maroon',linestyle='dashed',label='BOT')
plt.legend(loc='upper right')
plt.savefig('F1Plots/AlfaRomeoLapTimes2022Bahrain.png')
plt.show()

fig, ax = plt.subplots()
plt.suptitle(f"ALB vs LAT {race.weekend.name} {race.weekend.year}")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
ax.plot(alb['LapNumber'], alb['LapTime'], color='#89CFF0',label='ALB')
ax.plot(lat['LapNumber'], lat['LapTime'], color='#89CFF0',linestyle='dashed',label='LAT')
plt.legend(loc='upper right')
plt.savefig('F1Plots/WilliamsLapTimes2022Bahrain.png')
plt.show()

fig, ax = plt.subplots()
plt.suptitle(f"TSU vs GAS {race.weekend.name} {race.weekend.year}")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
ax.plot(tsu['LapNumber'], tsu['LapTime'], color='navy',label='TSU')
ax.plot(gas['LapNumber'], gas['LapTime'], color='navy',linestyle='dashed',label='GAS')
plt.legend(loc='upper right')
plt.savefig('F1Plots/AlphaTauriLapTimes2022Bahrain.png')
plt.show()

fig, ax = plt.subplots()
plt.suptitle(f"MAG vs MSC {race.weekend.name} {race.weekend.year}")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
ax.plot(mag['LapNumber'], mag['LapTime'], color='gray',label='MAG')
ax.plot(msc['LapNumber'], msc['LapTime'], color='gray',linestyle='dashed',label='MSC')
plt.legend(loc='upper right')
plt.savefig('F1Plots/HaasLapTimes2022Bahrain.png')
plt.show()

fig, ax = plt.subplots()
plt.suptitle(f"VET vs STR 2021 {race.weekend.name} {race.weekend.year}")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
ax.plot(sto['LapNumber'], sto['LapTime'], color='green',label='STR')
ax.plot(vet['LapNumber'], vet['LapTime'], color='green',linestyle='dashed',label='VET')
plt.legend(loc='upper right')
plt.savefig('F1Plots/AstonMartinLapTimes2022Bahrain.png')
plt.show()
```
